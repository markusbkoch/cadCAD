```python
import numpy as np
import pandas as pd
from copy import deepcopy

# genesis state
state_dict = {
    'a': 0,
    'b': 0,
    'state_history_df': None,
    'timestamp': '2019-01-01 00:00:00'
}

from datetime import timedelta
timestep_duration = timedelta(minutes=1) # In this example, a timestep has a duration of 1 minute.
from SimCAD.configuration.utils import ep_time_step

ts_format = '%Y-%m-%d %H:%M:%S'
def time_model(step, sL, s, _input):
    y = 'timestamp'
    x = ep_time_step(s, dt_str=s['timestamp'], fromat_str=ts_format, _timedelta=timestep_duration)
    return (y, x)

def updateA(step, sL, s, _input):
#    print(f"updateA > Timestep: {s['time_step']}; Mechstep: {s['mech_step']}; sL: {sL}")
    y = 'a'
    x = s['a'] + 1
    return (y, x)

def updateB(step, sL, s, _input):
#    print(f"updateB > Timestep: {s['time_step']}; Mechstep: {s['mech_step']}; sL: {sL}")
    y = 'b'
    try:
        x = s['a'] + s['state_history_df']['a'].sum()
    except:
        x = s['a']
    return (y, x)

def updateStateHistoryDF(step, sL, s, _input):
    y = 'state_history_df'
    sCopy = deepcopy(s)
    sCopy.pop('state_history_df')
    sCopy['mech_step'] = sCopy['mech_step'] + 1
    try:
        x = s['state_history_df'].append(pd.DataFrame([sCopy]))
    except:
        x = pd.DataFrame([sCopy])
    return (y, x)

partial_state_update_blocks = {
    'PSUB1': { 
        'behaviors': { 
        },
        'states': { 
            'state_history_df': updateStateHistoryDF,
            'timestamp': time_model,
            'a': updateA,
            'b': updateB
        }
    },
    'PSUB2': { 
        'behaviors': { 
        },
        'states': { 
            'state_history_df': updateStateHistoryDF,
            'a': updateA,
            'b': updateB
        }
    },
    'PSUB3': { 
        'behaviors': { 
        },
        'states': { 
            'state_history_df': updateStateHistoryDF,
            'a': updateA,
            'b': updateB

        }
    }
}

sim_config = {
    "N": 2,
    "T": range(3)
}

from SimCAD.configuration import Configuration
configs=[]
configs.append(Configuration(sim_config, state_dict, {}, {}, {}, partial_state_update_blocks))
from SimCAD.engine import ExecutionMode, ExecutionContext, Executor
exec_mode = ExecutionMode()
exec_context = ExecutionContext(exec_mode.single_proc)
executor = Executor(exec_context, configs) 
raw_result, tensor = executor.main()

import pandas as pd
pd.set_option('display.max_colwidth', 70)
df = pd.DataFrame(raw_result)    
df
```
